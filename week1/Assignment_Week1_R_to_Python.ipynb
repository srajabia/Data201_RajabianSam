## Part A -- Translation (Conceptual)

R code:
```{r}
x <- c(1, 3, 5, 7)
y <- ifelse(x > 4, x^2, x + 1)
mean(y)
```

1. Predict the output

**20**

2. Translate it into Python

```python
import numpy as np
x = np.array([1, 3, 5, 7])
y = np.where(x > 4, x^2, x + 1)
np.mean(y)
```

3. Compute the result in Python

````python
import numpy as np
x = np.array([1, 3, 5, 7])
y = np.where(x > 4, x**2, x + 1)
np.mean(y)
````

## Part B -- Lists vs NumPy Arrays

1. Create a Python list of numbers 1--5

```python
[1, 2, 3, 4, 5]
```
2. Attempt to multiply the list by 2

````python
[1, 2, 3, 4, 5] * 2
````

3. Convert the list to a NumPy array and repeat the multiplication

````python
import numpy as np
np.array([1, 2, 3, 4, 5]) * 2
````

- What happened in each case? <br>
**In the first case, the list repeated itself when multiplied by 2. In the second case, each element was multiplied by 2.**

- Why does NumPy behave differently? <br>
**NumPy converts the list into an array on which operations can be performed.**
